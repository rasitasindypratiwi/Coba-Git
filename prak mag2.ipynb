```python
import numpy as np
import pylab as plt

# Time interval
dt = 0.01
t = np.arange(-8, 8+dt, dt)
# Input signal x(t)
x = t
# Output signal y1(t) and y2(t)
y1 = np.sin(2*x*(t))
y2 = x * (t)+1

# Plot input signal
fig,ax = plt.subplots()
plt.plot(t,x)
plt.title("Input signal")
#Plot output signal
fig,ax = plt.subplots()
plt.plot(t,y1,label="$y_1(t)$")
plt.plot(t,y2,'--',label="$y_1(t)$")
plt.legend()
plt.title("Output signals")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# interval waktu
t = np.linspace(-8, 8, 1000)
x = t   # input utama

# ====================
# Sistem S1
# ====================
y1_S1 = np.sin(2 * x)    # y1(t) = sin(2x(t))
y2_S1 = x + 1            # y2(t) = x(t) + 1

# ====================
# Sistem S2
# input ke S2 adalah (y1_S1, y2_S1)
# ====================
y1_S2 = y1_S1 * y2_S1               # y1(t) = x1 * x2
y2_S2 = (y1_S1**2) - y2_S1          # y2(t) = (x1^2) - x2

# ====================
# Sistem S3
# input ke S3 adalah (y1_S2, y2_S2)
# ====================
y_S3 = np.where(t >= 0, y1_S2 + 2, -y2_S2)

# ====================
# Plot hasil
# ====================
plt.figure(figsize=(12,10))

# S1
plt.subplot(3,1,1)
plt.plot(t, y1_S1, label="y1(t) = sin(2t)")
plt.plot(t, y2_S1, label="y2(t) = t+1")
plt.title("Output Sistem S1")
plt.legend()
plt.grid(True)

# S2
plt.subplot(3,1,2)
plt.plot(t, y1_S2, label="y1(t) = y1_S1 * y2_S1")
plt.plot(t, y2_S2, label="y2(t) = (y1_S1^2) - y2_S1")
plt.title("Output Sistem S2")
plt.legend()
plt.grid(True)

# S3
plt.subplot(3,1,3)
plt.plot(t, y_S3, label="y(t) dari S3")
plt.title("Output Sistem S3")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
